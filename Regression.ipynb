# Linear Regression

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.rcParams['figure.figsize']=(20.0,10.0)

#for reading the data i.e. importing the data
data = pd.read_csv('Advertising.csv')
print(data.shape)
data.head(15)
```

```python
# Collecting the X and the Y values
X = data['TV'].values
Y = data['sales'].values
```

```python
# calculating the means of x and y
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Finding the total number of values in the dataset
n = len(X)

# calculating the b0 and b1 values using the formula

numer = 0
denom = 0

for i in range(n):
    numer += (X[i]-mean_x)*(Y[i]-mean_y)
    denom += (X[i]-mean_x)**2
    
b1 = numer/denom
b0 = mean_y - (b1*mean_x)

# Printing the values of b0 and b1
print(b1,b0)
```

# Visualization of the Regression Model 

```python
# Plotting the regression line 

max_x = np.max(X) + 100
min_x = np.min(X) - 100

# Calculating line values x and y
x = np.linspace(min_x,max_x,1000)
y = b0 + b1*x

# Plotting the line
plt.plot(x,y,color='blue', label='Regression line')
# Plotting the scatter points
plt.scatter(X,Y,c='red', label='Scatter points')

plt.xlabel('TV')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Calculating the R squared value now
r_numer = 0
r_denom = 0
for i in range(n):
    y_pred = b0 + b1*X[i]
    r_numer += (y_pred - mean_y)**2
    r_denom += (Y[i] - mean_y)**2
r2 = r_numer/r_denom
print(r2)
```

```python
# Thus we get the R2 value as 0.6118750508500699
print(r2)
```

# Multiple Regression

```python
from sklearn import linear_model
print(data.shape)
data.head(15)
```

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection="3d")

plt.show()
```

```python
# We'll perform multiple regression using linear_model

# creating object of the linear_model class
reg = linear_model.LinearRegression()
reg.fit(data[['TV','radio']],data.sales) #used to train the model
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# predicting the value of sales for any data of TV and radio

print('Calculating the data using the formula directly')
value = 2.9388893694594085 + 0.04576465*200 +  0.18853002*50
print(value)

print('Predicting the data using predefined module')
reg.predict([[200,50]])
```
